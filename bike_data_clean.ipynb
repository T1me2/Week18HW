```python
#Dpendencies and setup
import pandas as pd
import numpy as np

# Read in data
oct_2017_1 = pd.read_csv('resources/oct_2017.csv')
oct_2017_2 = pd.read_csv('resources/oct_2017_2.csv')
oct_2017_3 = pd.read_csv('resources/oct_2017_3.csv')
oct_2018_1 = pd.read_csv('resources/oct_2018.csv')
oct_2018_2 = pd.read_csv('resources/oct_2018_2.csv')
oct_2018_3 = pd.read_csv('resources/oct_2018_4.csv')
oct_2019_1 = pd.read_csv('resources/oct_2019.csv')
oct_2019_2 = pd.read_csv('resources/oct_2019_2.csv')
oct_2020_1 = pd.read_csv('resources/oct_2020.csv')
oct_2020_2 = pd.read_csv('resources/oct_2020_2.csv')
oct_2022_1 = pd.read_csv('resources/oct_2022.csv')
oct_2022_2 = pd.read_csv('resources/oct_2022_2.csv')
```

```python
oct_2020_2.dtypes
```

```python
oct_2022_2.dtypes
```

```python
#combine df's with the same column names/data types
oct_2017_2020 = pd.concat([oct_2017_1, oct_2017_2, oct_2017_3, oct_2018_1, oct_2018_2, oct_2018_3, oct_2019_1, oct_2019_2, oct_2020_1, oct_2020_2], ignore_index=True)
oct_2017_2020
```

```python
#rename column names in 2022 for columns that have matching data
oct_2022_1.rename(columns={
    'started_at': 'starttime',
    'ended_at': 'stoptime',
    'start_station_name': 'start station name',
    'end_station_name': 'end station name',
    'start_lat': 'start station latitude',
    'start_lng': 'start station longitude',
    'end_lat': 'end station latitude',
    'end_lng': 'end station longitude',
    'start_station_id': 'start station id',
    'end_station_id': 'end station id',
    'member_casual': 'usertype'
}, inplace = True)

oct_2022_2.rename(columns={
    'started_at': 'starttime',
    'ended_at': 'stoptime',
    'start_station_name': 'start station name',
    'end_station_name': 'end station name',
    'start_lat': 'start station latitude',
    'start_lng': 'start station longitude',
    'end_lat': 'end station latitude',
    'end_lng': 'end station longitude',
    'start_station_id': 'start station id',
    'end_station_id': 'end station id',
    'member_casual': 'usertype'
}, inplace = True)
```

```python
#change values of usertype in 2022 data to match prior iterations
for x in range(len(oct_2022_1)):
    if oct_2022_1.loc[x, 'usertype'] == 'member':
        oct_2022_1.loc[x, 'usertype'] = 'Subscriber'
    elif oct_2022_1.loc[x, 'usertype'] == 'casual':
        oct_2022_1.loc[x, 'usertype'] = 'Customer'
```

```python
#change values of usertype in 2022 data to match prior iterations
for x in range(len(oct_2022_2)):
    if oct_2022_2.loc[x, 'usertype'] == 'member':
        oct_2022_2.loc[x, 'usertype'] = 'Subscriber'
    elif oct_2022_2.loc[x, 'usertype'] == 'casual':
        oct_2022_2.loc[x, 'usertype'] = 'Customer'
oct_2022_2
```

```python
#drop columns that dont match up with old data
oct_2022_1 = oct_2022_1.drop(columns=['ride_id','rideable_type'])
oct_2022_2 = oct_2022_2.drop(columns=['ride_id','rideable_type'])
oct_2022_2
```

```python
#add in 2022 data
oct_2017_2022 = pd.concat([oct_2017_2020, oct_2022_1, oct_2022_2], ignore_index=True)
oct_2017_2022
```

```python
#save aggregated data
oct_2017_2022.to_csv('oct_2017_2022.csv')
```
